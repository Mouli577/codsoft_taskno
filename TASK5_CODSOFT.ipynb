```python
# ipython-input-65-c1e7fa5808d0
# create a function that seperates the class of transaction between fraud and legit transactions
def split_data_by_class(data):
    legit = data[data['Class'] == 0]
    fraud = data[data['Class'] == 1]
    return legit, fraud

# Call the function to create legit_df and fraud_df

legit_df, fraud_df = split_data_by_class(data)
```

```python
#sampling the legit transaction to match a 492 rows
new_legit_df = legit_df.sample(n = 492)
```

```python
print(new_legit_df)
```

```python
# combine the fraud and the fraud datasets
combined_df = pd.concat([new_legit_df, fraud_df], axis = 0)
```

```python
x = combined_df.drop('Class', axis = 1)
y = combined_df['Class']
```

```python
x
```

```python
y
```

```python
# importing the model building libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

SPLITTING AND TRAINING DATASETS

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 5)
```

```python
print(x_train)
```

```python
print(x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

MODEL EVALUATION

```python
# Testing the accuracy of the training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print(f"The model's training data accuracy is: {round(training_data_accuracy*100,2)}%")
```

```python
# Testing the accuracy of the training data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)
print(f"The model's testing data accuracy is: {round(testing_data_accuracy*100,2)}%")
```

```python

# Import necessary library
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

ax = ConfusionMatrixDisplay(confusion_matrix(y_test, x_test_prediction), display_labels = model.classes_)
ax.plot()
plt.show()
```
